## First examples

Make an object the last line of a cell to get the typeset display card.

```python
F = fd.foundation(fd.uniform_matroid(2, 4))
F
```

```python
F.show_hexagons()   # hexagons drawn as actual hexagons
```

```python
# the foundation of an elliptic matroid T_p, p = 11
CF = fd.canonical_foundation(fd.elliptic_matroid(11))
CF
```

```python
# morphisms = representations up to rescaling equivalence
fd.morphisms(fd.Pasture.from_foundation(F), fd.gf(5))
```

```python
# the reduced Dressian of T_11: a 1-dimensional fan with (p-1)/2 = 5 rays
D = fd.reduced_dressian(fd.elliptic_matroid(11))
D.f_vector(), sorted(D.rays())
```
