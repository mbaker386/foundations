# Foundations of matroids — runs entirely in your browser

Nothing to install: this page **is** a Python environment (Pyodide/WebAssembly).
Run the next cell once (Shift+Enter); the first run takes ~15 seconds while
Python and sympy load.

```python
try:
    import foundation as fd
except ImportError:
    from pyodide.http import pyfetch
    _r = await pyfetch('https://raw.githubusercontent.com/mbaker386/foundations/main/foundation.py')
    open('foundation.py', 'wb').write(await _r.bytes())
    import foundation as fd
print('foundation.py version:', fd.__version__)
```

## The foundation of a matroid, with typeset presentation

```python
F = fd.foundation(fd.uniform_matroid(2, 4))
F
```

```python
F.show_hexagons()
```

```python
CF = fd.canonical_foundation(fd.elliptic_matroid(11))
CF
```

```python
fd.morphisms(fd.Pasture.from_foundation(F), fd.gf(5))
```
